```python
import tensorflow as tf
import numpy as np
```

```python
kilogramos = np.array([1, 5, 10, 20, 50, 75, 100], dtype='float')
libras = np.array([2.20, 11.02, 22.05, 44.09, 110.23, 165.35, 220.46], dtype='float')
```

```python
oculta1 = tf.keras.layers.Dense(units=3, input_shape=[1])
oculta2 = tf.keras.layers.Dense(units=3)
oculta3 = tf.keras.layers.Dense(units=3)
salida = tf.keras.layers.Dense(units=1)
modelo = tf.keras.Sequential([oculta1, oculta2, oculta3, salida])
```

```python
modelo.compile(
    optimizer = tf.keras.optimizers.Adam(00.1),
    loss = 'mean_squared_error'

)
```

```python
print("Entrenando...")
historial = modelo.fit(kilogramos, libras, epochs=1000, verbose=False)
print("Fin de Entrenamiento...")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("#Epoca")
plt.ylabel("Magnitud de perdida")
plt.plot(historial.history['loss'])
```

```python
print("Prediccion.")
resultado = modelo.predict(np.array([150.0]))
print(f"El resultado es {resultado} L")
```
