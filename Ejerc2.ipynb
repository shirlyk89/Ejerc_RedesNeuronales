```python
import tensorflow as tf
import numpy as np
```

```python
import numpy as np
radio = np.array([1, 2, 3, 5, 8, 10, 12], dtype='float')
area = np.array([3.14, 12.56, 28.27, 78.53, 201.06, 314.15, 452.38], dtype='float')
```

```python
oculta1 = tf.keras.layers.Dense(64, activation='relu', input_shape=[1])
oculta2 = tf.keras.layers.Dense(64, activation='relu')
oculta3 = tf.keras.layers.Dense(32, activation='relu')
oculta4 = tf.keras.layers.Dense(16, activation='relu')
salida = tf.keras.layers.Dense(units=1)
modelo = tf.keras.Sequential([oculta1, oculta2, oculta3, salida])
```

```python
modelo.compile(
    optimizer = tf.keras.optimizers.Adam(0.1),
    loss = 'mean_squared_error'
)
```

```python
print("Entrenando...")
modelo.compile(
    optimizer = tf.keras.optimizers.Adam(0.001),
    loss = 'mean_squared_error'
)
historial = modelo.fit(radio, area, epochs=3000, verbose=False)
print("Fin de Entrenamiento...")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("#Epoca")
plt.ylabel("Magnitud de perdida")
plt.plot(historial.history['loss'])
```

```python
print("Prediccion.")
resultado =modelo.predict(np.array([15]))
print(f"El resultado es: {resultado}")
```
